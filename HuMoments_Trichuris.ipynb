# Hu Moments of the Images

```python

def HuMoments(images):
    
    f = np.asarray(images, dtype=np.float64)
    
    eta = np.zeros(shape= (4, 4), dtype = 'float64')
    
    for i in range(4):
        for j in range(4):
            eta[i, j] = ScaleInvariance(images, i, j)
    
    eta00 = eta[0, 0]
    eta01 = eta[0, 1]
    eta02 = eta[0, 2]
    eta03 = eta[0, 3]
    
    eta10 = eta[1, 0]
    eta11 = eta[1, 1]
    eta12 = eta[1, 2]
    eta13 = eta[1, 3]
    
    eta20 = eta[2, 0]
    eta21 = eta[2, 1]
    eta22 = eta[2, 2]
    eta23 = eta[2, 3]
    
    eta30 = eta[3, 0]
    eta31 = eta[3, 1]
    eta32 = eta[3, 2]
    eta33 = eta[3, 3]
   
   
   
    M1 = eta20 + eta02;
    M2 = (eta20 - eta02)**2 + (4*eta11**2);
    M3 = (eta30 - 3*eta12)**2 + (3*eta21 - eta03)**2;
    M4 = (eta30 + eta12)**2 + (eta21 + eta03)**2;
    M5 = (eta30 - 3*eta12) * (eta30 + eta12) * ((eta30 + eta12)**2 - (3*(eta21 + eta03)**2)) + (3*eta21 - eta03) * (eta21 + eta03) * ((3*(eta30 + eta12)**2) - (eta21 + eta03)**2);
    M6 = (eta20 - eta02)*((eta30 + eta12)**2 - (eta21 + eta03)**2) + 4*eta11*(eta30 + eta12)*(eta21 + eta03);
    M7 = (3*eta21 - eta03) * (eta30 + eta12) * ((eta30 + eta12)**2 - 3*(eta21 + eta03)**2) - (eta30 - 3*eta12) * (eta21 + eta03) * (3*(eta30 + eta12)**2 - (eta21 + eta03)**2);
    
    
   
    
    # The vector phi is a column vector containing M1,M2,....M7
    phi = [M1, M2, M3, M4, M5, M6, M7];
    
    
    return phi

def ScaleInvariance(images, p, q):
    f = np.asarray(images, dtype=np.float64)
    row = images.shape[0]
    col = images.shape[1]
    
    eta = 0
    mu00 = CentralMoment(images, 0, 0)
    mupq = CentralMoment(images, p, q)
    gama = (1+(p+q)/2)
    
    if p+q >= 2:
        eta = mupq/(mu00**gama)
        
    return eta

def CentralMoment(images, p, q):
    f = np.asarray(images, dtype=np.float64)
    row = images.shape[0]
    col = images.shape[1]
    
    m10 = Moment(images, 1, 0)
    m01 = Moment(images, 0, 1)
    m00 = Moment(images, 0, 0)
    
    x_bar = m10/m00
    y_bar = m01/m00
    
    mu = 0
    
    for i in range(row):
        for j in range(col):
            mu = mu + ((j - x_bar)**p)*((i - y_bar)**q)*f[i, j]
            
    return mu

def Moment(images, p, q):
    f = np.asarray(images, dtype=np.float64)
    #print(f.shape)
    
    row = images.shape[0]
    col = images.shape[1]
    
    m = 0
    for i in range(row):
        for j in range(col):
            m=m+((j**p)*(i**q))*f[i, j]
    return m

h = np.zeros(shape=(1000,7), dtype='float64') 
for i in range(len(loaded_gray)):
    hm = HuMoments(loaded_gray[i])  
#for i in range(1000):
    if i == 1000:
        break
    h[i][:]=hm
    print(i+1)
    print('--------------------------------------------------------------------------')
    print(h[i])
```
